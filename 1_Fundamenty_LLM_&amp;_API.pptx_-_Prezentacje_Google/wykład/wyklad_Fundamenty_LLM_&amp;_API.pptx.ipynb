````python
import pandas as pd
import kagglehub

rag_worksop = kagglehub.dataset_download('martininf1n1ty/rag-workshop')

df = pd.read_csv(f'{rag_worksop}/adresowo_offers_all_categories.csv')
df.head()
````

````python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

labels = ["sprzedaż domu", "sprzedaż mieszkania", "wynajem domu", "wynajem mieszkania"]

def predict_sentiment(text):
   if not text.strip():
       return "Wpisz zdanie przed klasyfikacją."
  
   result = classifier(text, candidate_labels=labels)
   label = result['labels'][0]
   score = result['scores'][0]
   return f"Sentyment: **{label}** (pewność: {score:.2%})"

````
